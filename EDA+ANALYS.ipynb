```python
# ==========================================
# üèÅ 1. –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫ –∏ –∑–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df = pd.read_csv('BMW sales data (2010-2024) .csv')

# –û–±—â–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è
print("–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞:", df.shape)
df.head()
```

```python
# ==========================================
# üîç 2. –ü–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –∏ –æ—á–∏—Å—Ç–∫–∞
# ==========================================

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–æ–ø—É—Å–∫–∏
df.info()
print("\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º:")
print(df.isnull().sum())

# –ü—Ä–æ–≤–µ—Ä–∏–º –¥—É–±–ª–∏–∫–∞—Ç—ã
print("\n–î—É–±–ª–∏–∫–∞—Ç–æ–≤ —Å—Ç—Ä–æ–∫:", df.duplicated().sum())

# –ü—Ä–æ–≤–µ—Ä–∏–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö
for col in ['Model', 'Region', 'Color', 'Fuel_Type', 'Transmission', 'Sales_Classification']:
    print(f"\n{col}: {df[col].nunique()} —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π")
    print(df[col].value_counts().head())
```

```python
# ==========================================
# üìä 3. –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
# ==========================================
df.describe().T

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–±—Ä–æ—Å –ø–æ –≥–æ–¥–∞–º
plt.figure(figsize=(10,4))
sns.countplot(data=df, x='Year', palette='Blues')
plt.title('üìÖ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–∂ –ø–æ –≥–æ–¥–∞–º')
plt.xticks(rotation=45)
plt.show()
```

```python
# ==========================================
# üöó 4. –ê–Ω–∞–ª–∏–∑ –ø–æ –º–æ–¥–µ–ª—è–º –∏ —Ä–µ–≥–∏–æ–Ω–∞–º
# ==========================================

# –¢–æ–ø-10 –º–æ–¥–µ–ª–µ–π –ø–æ –æ–±—ä—ë–º—É –ø—Ä–æ–¥–∞–∂
top_models = df.groupby('Model')['Sales_Volume'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_models.values, y=top_models.index, palette='viridis')
plt.title('üî• –¢–æ–ø-10 –º–æ–¥–µ–ª–µ–π –ø–æ –æ–±—ä—ë–º—É –ø—Ä–æ–¥–∞–∂')
plt.xlabel('Sales Volume')
plt.show()

# –ü—Ä–æ–¥–∞–∂–∏ –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Region', y='Sales_Volume', estimator=sum, ci=None, palette='coolwarm')
plt.title('üåç –ü—Ä–æ–¥–∞–∂–∏ –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º')
plt.xticks(rotation=30)
plt.show()
```

```python
# ==========================================
# üí∞ 5. –ê–Ω–∞–ª–∏–∑ —Ü–µ–Ω –∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
# ==========================================

# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–Ω
plt.figure(figsize=(8,5))
sns.histplot(df['Price_USD'], bins=30, kde=True, color='skyblue')
plt.title('üíµ –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–Ω –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–π')
plt.show()

# –°–≤—è–∑—å –º–µ–∂–¥—É —Ü–µ–Ω–æ–π –∏ –æ–±—ä—ë–º–æ–º –¥–≤–∏–≥–∞—Ç–µ–ª—è
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Engine_Size_L', y='Price_USD', hue='Fuel_Type', alpha=0.7)
plt.title('‚öôÔ∏è –û–±—ä—ë–º –¥–≤–∏–≥–∞—Ç–µ–ª—è vs. –¶–µ–Ω–∞')
plt.show()

# –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('üìà –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤')
plt.show()
```

```python
# ==========================================
# üí° 6. –ü—Ä–æ–¥—É–∫—Ç–æ–≤—ã–µ –≥–∏–ø–æ—Ç–µ–∑—ã
# ==========================================

print("""
üéØ –í–æ–∑–º–æ–∂–Ω—ã–µ –≥–∏–ø–æ—Ç–µ–∑—ã –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏:

1Ô∏è‚É£ –ú–æ–¥–µ–ª–∏ —Å –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–π —Ç—Ä–∞–Ω—Å–º–∏—Å—Å–∏–µ–π –∏–º–µ—é—Ç –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–µ –ø—Ä–æ–¥–∞–∂–∏, –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ—Å–ª–µ 2020 –≥–æ–¥–∞.
2Ô∏è‚É£ –≠–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–∏–µ –∏ –≥–∏–±—Ä–∏–¥–Ω—ã–µ –º–æ–¥–µ–ª–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç —Ä–æ—Å—Ç –ø—Ä–æ–¥–∞–∂ –≤ –ï–≤—Ä–æ–ø–µ –∏ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ.
3Ô∏è‚É£ –ë–æ–ª–µ–µ –º–æ—â–Ω—ã–µ –¥–≤–∏–≥–∞—Ç–µ–ª–∏ (Engine_Size_L > 3.0) –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—Ç —Å —Ü–µ–Ω–æ–π, –Ω–æ –Ω–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å –æ–±—ä—ë–º–æ–º –ø—Ä–æ–¥–∞–∂.
4Ô∏è‚É£ –ß–µ—Ä–Ω—ã–π –∏ —Å–µ—Ä—ã–π —Ü–≤–µ—Ç–∞ —á–∞—â–µ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –≤ –ø—Ä–µ–º–∏—É–º-—Å–µ–≥–º–µ–Ω—Ç–µ (High Sales_Classification).
5Ô∏è‚É£ –ü–æ—Å–ª–µ 2020 –≥–æ–¥–∞ –ø—Ä–æ–¥–∞–∂–∏ —Ä–∞—Å—Ç—É—Ç –±—ã—Å—Ç—Ä–µ–µ –≤ —Ä–µ–≥–∏–æ–Ω–∞—Ö —Å –≥–∏–±—Ä–∏–¥–Ω—ã–º–∏ –º–æ–¥–µ–ª—è–º–∏ (–≤–ª–∏—è–Ω–∏–µ —ç–∫–æ—Ç—Ä–µ–Ω–¥–æ–≤).
""")
```

```python
# ==========================================
# üß† 7. –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–¥—É–∫—Ç–æ–≤—ã—Ö –≥–∏–ø–æ—Ç–µ–∑
# ==========================================
import scipy.stats as stats

# ------------------------------------------
# –ì–∏–ø–æ—Ç–µ–∑–∞ 1:
# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è —Ç—Ä–∞–Ω—Å–º–∏—Å—Å–∏—è -> –≤—ã—à–µ –æ–±—ä—ë–º –ø—Ä–æ–¥–∞–∂, –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ—Å–ª–µ 2020 –≥–æ–¥–∞
# ------------------------------------------
plt.figure(figsize=(8,5))
sns.boxplot(data=df[df['Year'] >= 2020], x='Transmission', y='Sales_Volume', palette='Set2')
plt.title('‚öôÔ∏è –ü—Ä–æ–¥–∞–∂–∏ –ø–æ —Ç–∏–ø—É —Ç—Ä–∞–Ω—Å–º–∏—Å—Å–∏–∏ (2020+ –≥–æ–¥—ã)')
plt.show()

auto_sales = df[(df['Transmission'] == 'Automatic') & (df['Year'] >= 2020)]['Sales_Volume']
manual_sales = df[(df['Transmission'] == 'Manual') & (df['Year'] >= 2020)]['Sales_Volume']

t_stat, p_value = stats.ttest_ind(auto_sales, manual_sales, equal_var=False)
print(f'P-value = {p_value:.4f}')
if p_value < 0.05:
    print("‚úÖ –†–∞–∑–ª–∏—á–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã ‚Äî –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –º–æ–¥–µ–ª–∏ –ø—Ä–æ–¥–∞—é—Ç—Å—è –ª—É—á—à–µ –ø–æ—Å–ª–µ 2020 –≥–æ–¥–∞.")
else:
    print("‚ùå –ù–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –º–µ–∂–¥—É —Ç–∏–ø–∞–º–∏ —Ç—Ä–∞–Ω—Å–º–∏—Å—Å–∏–∏.")


# ------------------------------------------
# –ì–∏–ø–æ—Ç–µ–∑–∞ 2:
# –ì–∏–±—Ä–∏–¥–Ω—ã–µ –∏ —ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–∏–µ –º–æ–¥–µ–ª–∏ —Ä–∞—Å—Ç—É—Ç –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º –±—ã—Å—Ç—Ä–µ–µ –¥—Ä—É–≥–∏—Ö —Ç–∏–ø–æ–≤ —Ç–æ–ø–ª–∏–≤–∞
# ------------------------------------------
fuel_trends = df.groupby(['Year', 'Fuel_Type'])['Sales_Volume'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=fuel_trends, x='Year', y='Sales_Volume', hue='Fuel_Type', marker='o')
plt.title('üîã –î–∏–Ω–∞–º–∏–∫–∞ –ø—Ä–æ–¥–∞–∂ –ø–æ —Ç–∏–ø—É —Ç–æ–ø–ª–∏–≤–∞ (2010‚Äì2024)')
plt.show()


# ------------------------------------------
# –ì–∏–ø–æ—Ç–µ–∑–∞ 3:
# –ë–æ–ª—å—à–∏–π –æ–±—ä—ë–º –¥–≤–∏–≥–∞—Ç–µ–ª—è —Å–≤—è–∑–∞–Ω —Å –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–æ–π —Ü–µ–Ω–æ–π, –Ω–æ –Ω–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å –±–æ–ª—å—à–∏–º –æ–±—ä—ë–º–æ–º –ø—Ä–æ–¥–∞–∂
# ------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(14,5))

sns.scatterplot(data=df, x='Engine_Size_L', y='Price_USD', alpha=0.6, ax=axes[0])
axes[0].set_title('‚öôÔ∏è –û–±—ä—ë–º –¥–≤–∏–≥–∞—Ç–µ–ª—è vs. –¶–µ–Ω–∞')

sns.scatterplot(data=df, x='Engine_Size_L', y='Sales_Volume', alpha=0.6, ax=axes[1])
axes[1].set_title('‚öôÔ∏è –û–±—ä—ë–º –¥–≤–∏–≥–∞—Ç–µ–ª—è vs. –û–±—ä—ë–º –ø—Ä–æ–¥–∞–∂')

plt.show()

corr_engine_price = df['Engine_Size_L'].corr(df['Price_USD'])
corr_engine_sales = df['Engine_Size_L'].corr(df['Sales_Volume'])
print(f"üìà –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è Engine_Size ‚Üî Price: {corr_engine_price:.2f}")
print(f"üìâ –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è Engine_Size ‚Üî Sales_Volume: {corr_engine_sales:.2f}")


# ------------------------------------------
# –ì–∏–ø–æ—Ç–µ–∑–∞ 4:
# –¢—ë–º–Ω—ã–µ —Ü–≤–µ—Ç–∞ (Black, Grey, Silver) —á–∞—â–µ –ø–æ–ø–∞–¥–∞—é—Ç –≤ High Sales Classification
# ------------------------------------------
df['Is_Dark_Color'] = df['Color'].str.lower().isin(['black', 'grey', 'gray', 'silver'])
color_class_pivot = pd.crosstab(df['Is_Dark_Color'], df['Sales_Classification'], normalize='index') * 100
color_class_pivot.plot(kind='bar', stacked=True, figsize=(6,4), colormap='coolwarm')
plt.title('üé® –î–æ–ª—è High/Low –ø—Ä–æ–¥–∞–∂ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ü–≤–µ—Ç–∞')
plt.show()


# ------------------------------------------
# –ì–∏–ø–æ—Ç–µ–∑–∞ 5:
# –ü–æ—Å–ª–µ 2020 –≥–æ–¥–∞ —Ä–µ–≥–∏–æ–Ω—ã —Å –≥–∏–±—Ä–∏–¥–Ω—ã–º–∏ –º–æ–¥–µ–ª—è–º–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –±–æ–ª—å—à–∏–π —Ä–æ—Å—Ç –ø—Ä–æ–¥–∞–∂
# ------------------------------------------
hybrid_growth = (
    df[df['Fuel_Type'] == 'Hybrid']
    .groupby(['Year', 'Region'])['Sales_Volume']
    .sum()
    .reset_index()
)

plt.figure(figsize=(10,5))
sns.lineplot(data=hybrid_growth, x='Year', y='Sales_Volume', hue='Region', marker='o')
plt.title('üåç –ü—Ä–æ–¥–∞–∂–∏ –≥–∏–±—Ä–∏–¥–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º (2010‚Äì2024)')
plt.show()
```
